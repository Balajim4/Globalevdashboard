```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

print("🚀 Level 2.2: Deep Scanning EV Database Grid...")

url = "https://ev-database.org/"
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Intha site-la ovvoru bike/car-um 'data-row' la irukkum
    cars = soup.find_all('div', class_='list-item')

    ev_data = []

    for car in cars:
        try:
            # Extracting Name
            title = car.find('div', class_='title-wrap').text.strip()

            # Extracting Specs (Battery, Range, etc.)
            battery = car.find('span', class_='battery').text.strip()
            acceleration = car.find('span', class_='acceleration').text.strip()

            ev_data.append({
                "Model": title,
                "Battery": battery,
                "0-100": acceleration
            })
        except AttributeError:
            continue # Edhavathu miss aana skip pannu

    # DataFrame-a maathurom
    df = pd.DataFrame(ev_data)
    print(f"✅ Success! Captured {len(df)} EV Models.")
    print(df.head())

    df.to_csv("real_v2_clean.csv", index=False)
    print("\n💾 Data saved! Dashboard ready mode.")
else:
    print("❌ Connection issue.")
  # 'Model' column-la irukka double text-a clean panna oru simple logic
df['Model'] = df['Model'].apply(lambda x: x.split('\n')[0].strip())

# Ippo paaru, data 'Tesla Model 3' nu clean-a irukkum!
print(df.head())
```

```python
import pandas as pd

# Namma nethu/morning extract panna data-va load panrom
df = pd.read_csv("real_v2_clean.csv")

print("🧹 Starting Data Cleaning Surgery...")

# 1. Model Name-la irukka double text and '\n' thookuroam
df['Model'] = df['Model'].apply(lambda x: str(x).split('\n')[0].strip())

# 2. Battery and 0-100 columns-a numerical-a mathuroam (plotting-ku help aagum)
# 'kWh' or 'sec' mathiri characters-a clean panna
df['Battery_kWh'] = pd.to_numeric(df['Battery'], errors='coerce')
df['Accel_0_100_sec'] = pd.to_numeric(df['0-100'], errors='coerce')

# 3. Unwanted original columns-a drop pannalam
df_final = df[['Model', 'Battery_kWh', 'Accel_0_100_sec']].dropna()

print(f"✨ Cleaning Done! {len(df_final)} high-quality records ready.")
print(df_final.head())

# Dashboard-kku puthu file-a save panrom
df_final.to_csv("v2_dashboard_ready.csv", index=False)
print("\n✅ 'v2_dashboard_ready.csv' generated. Upload this to Streamlit!")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(page_title="Global EV Dashboard V2", layout="wide")

st.title("⚡ Global EV Database V2.0")
st.write("Extracting data from 1,200+ Global EV Models")

# Loading the cleaned data
df = pd.read_csv("v2_dashboard_ready.csv")

# --- CLEANING THE DATA FOR PLOTTING ---
# Battery and Accel columns-la empty values (NaN) iruntha thookiduvom
df = df.dropna(subset=['Battery_kWh', 'Accel_0_100_sec'])

# Metrics summary
col1, col2, col3 = st.columns(3)
col1.metric("Total Models", len(df))
col2.metric("Avg Battery (kWh)", f"{df['Battery_kWh'].mean():.1f}")
col3.metric("Avg 0-100 (sec)", f"{df['Accel_0_100_sec'].mean():.1f}")

# The Graph with Dark Mode & Clean Render
st.subheader("Battery Capacity vs Acceleration (0-100 km/h)")
fig = px.scatter(df, x="Battery_kWh", y="Accel_0_100_sec",
                 hover_name="Model",
                 color="Battery_kWh",
                 color_continuous_scale="Viridis",
                 template="plotly_dark", # Ithu thaan antha mass look!
                 title="Performance Analysis (1200+ Models)")

st.plotly_chart(fig, use_container_width=True)

# Data Table at the bottom
st.subheader("Raw Data View")
st.dataframe(df)
```
